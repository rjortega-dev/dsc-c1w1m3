# Practice for Join Statements with SQL

Import the necessary packages and connect to the database 'data.sqlite'.

```python
# Run without change

# Libraries and database

import sqlite3
import pandas as pd

conn = sqlite3.connect('data.sqlite')
```

1. Select the names (frist and last) of all employees in Boston

```python
# Your code here
q = """
SELECT firstName, lastName
	FROM employees
JOIN offices
USING(officeCode)
WHERE city = "Boston";
"""
pd.read_sql(q, conn)
```

2. Are there any offices that have zero employees?

```python
# Your code here
q = """
SELECT
	o.officeCode,
	o.city,
	COUNT(e.employeeNumber) AS employeesCount
FROM offices AS o
LEFT JOIN employees AS e
	USING(officeCode)
GROUP BY officeCode
HAVING employeesCount = 0;
"""
pd.read_sql(q, conn)
```

3. How many customers are there per office?

```python
# Your code here
q = """
SELECT
    o.officeCode,
    o.city,
    COUNT(c.customerNumber) AS customersCount
FROM offices AS o
JOIN employees AS e
    USING(officeCode)
JOIN customers AS c
    ON e.employeeNumber = c.salesRepEmployeeNumber
GROUP BY officeCode
;
"""
pd.read_sql(q, conn)
```

4. Display the names of every individual product that each employee has sold as a dataframe.

```python
# Your code here
q = """
SELECT firstName, lastName, productName
FROM employees AS e
JOIN customers AS c
    ON e.employeeNumber = c.salesRepEmployeeNumber
JOIN orders
    USING(customerNumber)
JOIN orderdetails
    USING(orderNumber)
JOIN products
    USING(productCode);
"""
df = pd.read_sql(q, conn)
df.head(10)
```

5. Display the number of products each employee has sold
- Alphabetize the results by employee last name.
- Use the quantityOrdered column from orderDetails.
- Think about how to group the data when some employees might have the same first or last name.


```python
# Your code here
q = """
SELECT firstName, lastName, SUM(quantityOrdered) as total_products_sold
FROM employees AS e
JOIN customers AS c
    ON e.employeeNumber = c.salesRepEmployeeNumber
JOIN orders
    USING(customerNumber)
JOIN orderdetails
    USING(orderNumber)
GROUP BY firstName, lastName
ORDER BY lastName;
"""
pd.read_sql(q, conn)
```

6.  Display the names of employees who have sold more than 200 different products.

```python
# Your codde here
q = """
SELECT firstName, lastName, COUNT(productCode) as different_products_sold
FROM employees AS e
JOIN customers AS c
    ON e.employeeNumber = c.salesRepEmployeeNumber
JOIN orders
    USING(customerNumber)
JOIN orderdetails
    USING(orderNumber)
GROUP BY firstName, lastName
HAVING different_products_sold > 200
ORDER BY lastName;
"""
pd.read_sql(q, conn)
```
